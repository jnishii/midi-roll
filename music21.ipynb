```python
target ="BWV846"
path="data/pedb2_v0.0.1.b/"
targetxml="bac-wtc101-p-a-p1/bac-wtc101-p.xml"
targetmid="bac-wtc101-p-a-p1/bac-wtc101-p-a-p1.mid"

piecemid=converter.parse(path+targetmid)
piecexml=converter.parse(path+targetxml)
# piece=piecemid analyze using midi
piece = piecexml # analyze using xml
```

```python
# use music21 corpus data

target ="BWV846"
piecexml = corpus.parse('bwv846')
piece = piecexml # analyze using xml
```

```python
piece.measures(13,16).show()
#piece.measures(14,18).show()
```
